```python
# ---------------------------------------------
# STEP 1: IMPORT LIBRARIES
# ---------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# ---------------------------------------------
# STEP 2: LOAD THE DATASET
# ---------------------------------------------
# Make sure 'diabetes.csv' is in your working directory
df = pd.read_csv("C://D_Drive//CSE//Data_Analyst_Project//Diabetes Dataset//diabetes.csv")

# Preview data
print(df.head())
print(df.info())
```

```python
# ---------------------------------------------
# STEP 3: DATA CLEANING
# ---------------------------------------------
# Columns where 0 is not a valid physiological value
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)

# Replace missing values with the median of the column
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# ---------------------------------------------
# STEP 4: DATA VISUALIZATION
# ---------------------------------------------
# Correlation heatmap to understand relationships
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")

# Save the figure as a .png file
plt.savefig('correlation_heatmap.png', bbox_inches='tight')  # Save with tight bounding box

plt.show()

```

```python
# ---------------------------------------------
# STEP 5: PREPARE FEATURES AND TARGET
# ---------------------------------------------
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# ---------------------------------------------
# STEP 6: TRAIN/TEST SPLIT
# ---------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

```python
# ---------------------------------------------
# STEP 7: MODEL TRAINING
# ---------------------------------------------
# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)

# Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

```python
# ---------------------------------------------
# STEP 8: EVALUATE RANDOM FOREST MODEL
# ---------------------------------------------
y_pred = rf.predict(X_test)

print("Classification Report:\n")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)

# Confusion Matrix Plot
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

# Save the confusion matrix plot
plt.savefig('confusion_matrix.png', bbox_inches='tight')

plt.show()
```

```python
# ---------------------------------------------
# STEP 9: FEATURE IMPORTANCE
# ---------------------------------------------
importances = rf.feature_importances_
feature_names = X.columns
sorted_idx = np.argsort(importances)

plt.figure(figsize=(10, 6))
plt.barh(feature_names[sorted_idx], importances[sorted_idx])
plt.xlabel("Feature Importance")
plt.title("Random Forest Feature Importance")

# Save the Feature Importance
plt.savefig('Feature_Importance.png', bbox_inches='tight')

plt.show()
```

```python
# Train Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# Evaluate Logistic Regression Model
y_pred_lr = lr.predict(X_test)

print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import roc_curve, auc

# Compute the ROC curve and AUC for Logistic Regression
fpr, tpr, thresholds = roc_curve(y_test, lr.predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='Logistic Regression (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line (random model)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve (Logistic Regression)')
plt.legend(loc='lower right')

# Save the ROC curve plot
plt.savefig('roc_curve.png', bbox_inches='tight')

plt.show()
```
